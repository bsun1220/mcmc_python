```python
from DikinWalk import DikinWalk
from VaidyaWalk import VaidyaWalk
from DikinLSWalk import DikinLSWalk
from JohnWalk import JohnWalk
from HitAndRun import HitAndRun
from BallWalk import BallWalk

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import sqrtm
from scipy.optimize import minimize, LinearConstraint, linprog
import pandas as pd
import sympy
```

```python
#Unit Square
A = np.matrix([[1, 0], [-1, 0],[0, 1], [0, -1]])
b = np.matrix([[1],[0],[1],[0]])
x = np.matrix([[0.5], [0.5]]) 
```

```python
#Unit Diamond
A = np.matrix([[1, 1], [-1, 1],[1, -1], [-1, -1]])
b = np.matrix([[1],[1],[1], [1]])
x = np.matrix([[0], [0]]) 
```

```python
#Right Triangle
A = np.matrix([[-1, 0], [0, -1],[1, 1]])
b = np.matrix([[0],[0],[1]])
x = np.matrix([[0.2], [0.3]]) 
```

```python
dikin = DikinWalk(A, b, x, 0.4)
results = dikin.generate_complete_walk(10000)
plt.scatter(results[:,0],results[:,1], alpha = 0.2)
plt.show()
```

```python
vaidya = VaidyaWalk(A, b, x, 0.4)
results = vaidya.generate_complete_walk(10)
plt.scatter(results[:,0],results[:,1], alpha = 0.2)
plt.show()
```

```python
dl = DikinLSWalk(A, b, x, 0.4)
results = dl.generate_complete_walk(10)
plt.scatter(results[:,0],results[:,1], alpha = 0.2)
plt.show()
```

```python
john = JohnWalk(A, b, x, 0.4)
results = john.generate_complete_walk(10000)
plt.scatter(results[:,0],results[:,1], alpha = 0.2)
plt.show()
```

```python
dikinls = DikinLSWalk(A, b, x, 0.4)
results = dikinls.generate_complete_walk(10000)
plt.scatter(results[:,0],results[:,1], alpha = 0.2)
plt.show()
```

```python
bwalk = BallWalk(A, b, 0.1)
results = bwalk.random_sample(x, 10000)
plt.scatter(results[:,0],results[:,1], alpha = 0.2)
plt.show()
```

```python
def hitting_time_metric(result, dist):
    dim = result.shape[1]
    df = pd.DataFrame(result)
    edge = np.random.randint(2, size = dim)
    remain = df[np.linalg.norm((df - edge), axis = 1) < dist]
    return -1 if len(remain) == 0 else remain.index[0]
```
